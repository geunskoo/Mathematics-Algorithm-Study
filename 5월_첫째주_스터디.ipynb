## DFS와 BFS

```python
n, m ,start = map(int, input().split()) # n = 정점의 개수 m = 간선의 개수 start = 시작할 정점

graph = {} # 그래프 그리는것을 dictionary를 이용
for i in range(1,n+1):
  graph[i] = [i]  # 딕셔너리 생성 
for i in range(m): # vertex들에 연결되는 간선 표현
  x,y = map(int, input().split())
  graph[x] += [y]
  graph[y] += [x]
for i in graph: # 방문할 수 있는 정점이 여러개인 경우 -> 정점 번호가 작은 것을 먼저 방문
  graph[i].sort()
for i in dfs(start, graph):# dfs함수를 이용하여 list를 return
  print(i, end = ' ') # list의 원소 하나씩 출력
print() # 한칸 내려서 출력 하기 위함
for j in bfs(start, graph): # bfs함수를 이용하여 list를 return
  print(j, end = ' ') # list의 원소 하나씩 출력
```

```python
def dfs(start, graph):
  from collections import deque # deque이용
  visited = deque() # 방문 여부 확인
  stack = deque([start]) 
  while stack:
    num = stack.popleft() # stack의 왼쪽 원소 하나 pop이용 <- deque 이용
    if num not in visited: # 방문 한 적 없을 경우
      visited.append(num) # 방문 처리
      stack = deque(graph[num] + list(stack)) # DFS는 끝까지 가야 하기 때문에
                                              # 왼쪽에 LIST를 추가(LIFO)
  return visited
```

```python
def bfs(start, graph):
  from collections import deque 
  visited = deque() # 방문 여부 처리
  queue = deque([start]) 
  while queue:
    num = queue.popleft() # QUEUE의 왼쪽 원소 하나 POP
    if num not in visited:# 방문 한 적 없는 경우
      visited.append(num) # 방문 처리
      queue.extend(graph[num]) # BFS는 하나 하나씩 확인 (FIFO)
  return visited
```

## 유기농 배추
1012번

```python
t = int(input()) # t = test 횟수
from collections import deque # deque 이용
dx = [-1,1,0,0] # 위 아래 좌 우 처리용
dy = [0,0,-1,1]


for _ in range(t):
  visited = [] # visit처리는 테스트별로 따로 처리
  answer = 0 # 필요한 지렁이의 개수
  w, h, c = map(int, input().split()) # 가로 세로는 반전되어도 일정하게만 사용하면 상관 x
  field = [] # 배추밭 생성
  for i in range(w):
    field.append([0]*h) # 우선 배추밭에 배추 없는 경우
  for _ in range(c):
    x, y = map(int, input().split())
    field[x][y] = 1 # 하나씩 받아서 배추 심어주기
  for i in range(w): # w*h 크기의 배추밭에서 함수처리
    for j in range(h):
      if field[i][j]==1 and [i,j] not in visited:
        if bfs_cabbage(i,j):
          answer += 1
  print(answer)
```

```python
def bfs_cabbage(start_x, start_y):
 
  start = [start_x, start_y] # start할 x,y 좌표
  stack = deque([start]) # 시작점 stack에 넣어주기 (뒤에서 while문 사용할 예정)
  visited.append(start) # visit 처리 (함수별로 visit을 따로 하면 안된다)
  count = 1
  while stack:
    x,y = stack.popleft() # bfs와 동일
    for i in range(4):
      nx = x + dx[i] # 상하좌우 확인
      ny = y + dy[i]
      if 0<=nx<w and 0<=ny<h: # index error 일어나지 않도록
        if field[nx][ny] == 1 and [nx, ny] not in visited: # 배추가 있고 방문한 적 없을 경우
          visited.append([nx, ny]) # 방문 처리
          stack.append([nx,ny]) # stack에 넣어서 해당 위치 상하좌우 확인
          count+=1
  return count # count는 무조건 1이상이므로 문제에선 bfs가 몇번 돌았는가만 생각하면 됨
```

```python
def selection_sort(list1):
  n = len(list1)
  for i in range(n):
    min = list1[i] 
    ind = i
    for j in range(i, n): # 최솟값 찾아서
      if min>list1[j]:
        min = list1[j]
        ind = j
    if i!=ind: # 최솟값과 바꿔주기
      list1[i], list1[ind] = list1[ind], list1[i]
  return
```

```python
def insert_sort(list1):
  n = len(list1)
  for i in range(1,n):
    num = list1[i]
    ind = i
    for j in range(i-1, -1,-1):
      if num<list1[j]:
        list1[ind], list1[j] = list1[j], list1[ind]
        ind = j
      else: break #뒤부터 확인 하므로 braek를 사용하여 시간을 좀 줄이자
  return
```

## 버블정렬
맨앞부터 하나씩 자기 자신의 오른쪽과 비교해서 정렬

```python
def bubble_sort(list1):
  num = len(list1)
  for i in range(num-1,0,-1):
    is_swapped = False
    for j in range(i):
      if list1[j]>list1[j+1]:
        list1[j], list1[j+1] = list1[j+1], list1[j]
        is_swapped = True
    if is_swapped==False: #한번 다 돌렸는데 변화가 없으면 
      break                   #모두 정렬된 것이므로 break
  return
```

## 퀵 정렬

피벗(pivot) 지정 후 pivot을 기준으로
비균등분할을 한다(merge sort는 균등분할)

pivot을 맨왼쪽, 맨오른쪽으로 고르기 (2가지)


```python
# pivot을 맨왼쪽에서 고르는 경우
def quick_sort1(data, s, e):
  if s>=e:
    return
  else:
    pivot = s # 피벗은 리스트릐 맨 왼쪽 원소
    i = s + 1 # 피벗 바로 오른쪽 숫자 (방향 ->)
    j = e # 맨 마지막 숫자 (방향 <-)
    while i<=j:
      while data[pivot]>=data[i]: # 피벗보다 큰 수가 나올때까지 i증가
        i+=1
        if i>e: # 인덱스 고려
          break
      while data[pivot]<=data[j]: # 피벗보다 작은 수가 나올때까지 1감소
        j-=1
        if j==s: # 인덱스 고려
          break
      if i>=j:
        data[pivot], data[j] = data[j], data[pivot] # 만약 i와 j의 인덱스 크기가 바뀌
                                                    # 면 피벗과 j의 원소 swap
      else:
        data[j], data[i] = data[i], data[j] # 아니면 그냥 i,j 의 원소 swap
    quick_sort1(data, s, j-1) # pivot 기준 왼쪽것 재귀
    quick_sort1(data, j+1, e) # pivot 기존 오른쪽 재귀
```

```python
# pivot을 맨오른쪽에서 고르는 경우
def quick_sort2(data, s, e):
  if s>=e:
    return
  else:
    pivot = e #pivot은 리스트 맨 오른쪽 원소
    i = s # 맨 왼쪽 원소
    j = e-1 # pivot 바로 왼쪽 원소
    while i<j:
      while data[pivot]>=data[i]: # 피벗보다 큰 수 나올때까지 i증가
        i+=1
        if i==e: #인덱스 고려
          break
      while data[pivot]<=data[j]: # 피벗보다 작은 수 나올때까지 j감소
        j-=1
        if j<s: # 인덱스 고려
          break
      if i>j:
        data[pivot], data[i] = data[i], data[pivot]# 만약 i와 j의 인덱스 크기가 바뀌
        break                                           # 면 피벗과 j의 원소 swap
      else:
        data[j], data[i] = data[i], data[j] # 아니면 i,j의 원소 swap

    if data[pivot]<data[i]: # i=j인경우 오류 없애주기위함 56 52인데 안바뀌었다(52, 78 49 56 68)
      data[pivot], data[i] = data[i], data[pivot]
    quick_sort2(data, s, i-1)
    quick_sort2(data, i+1, e)
```

```python
def quick(a):
  if len(a)<=1:
    return a
  else:
    pivot = a[0]
    tail = a[1:]
    left = [x for x in tail if x<=pivot]
    right = [x for x in tail if x>pivot]
  return quick(left)+[pivot]+quick(right)
```

## 계수정렬
배열의 인덱스를 이용하여 데이터를 저장 할 것이다.


데이터가 양수여야 사용 가능

값의 범위가 너무 크면 사용을 할 수 없다.

```python
def counting_sort(list1):
  max_num = max(list1)
  count = [0]*(max_num+1) # 0부터 최댓값까지의 숫자개수 저장하는 리스트
  for i in list1:
    count[i] += 1 # 각 숫자 개수 저장
  answer = []
  for i in range(len(count)):
    if count[i]>0:
      for j in range(count[i]): # 각 숫자 개수만큼 answer 리스트에 저장
        answer.append(i)
  return answer
```

```python
n = int(input())
list1 = []
for i in range(n):
  list1.append(randint(0,10000))
print(list1)

start = time()
list1.sort() # 그냥 sort 사용
end = time()

print(list1)
print(f'{end-start:.5f}초 걸렸습니다')
# 10000에 약 0.00166초정도 걸린다!?
```
